But, the best way to predict whether one has diabetes or not is to check their insulin level and their glucose level.

If one's glocose level is higher than his/her insulin level, he/she has diabetes.

Q) How does the insulin level vary with respect to glucose levels in indivisuals with and without diabates? Using the given insulin and glucose levels in the dataset, predict whether is diabetic or not.

```python
plt.figure(figsize=(20, 20))
sns.scatterplot(x='Glucose', y='Insulin', hue='Outcome', data=df, palette='coolwarm')
plt.title('Scatter plot of Insulin vs. Glucose by Diabetes Status')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.legend(title='Diabetes Status', loc='upper left', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# This is to categorise glucose levels
df['Glucose_Category'] = pd.cut(df['Glucose'], bins=[0, 100, 140, df['Glucose'].max()], labels=['Low', 'Medium', 'High'])


plt.figure(figsize=(20, 20))
sns.boxplot(x='Glucose_Category', y='Insulin', hue='Outcome', data=df, palette='plasma')
plt.title('Box plot of Insulin by Glucose Category and Diabetes Status')
plt.xlabel('Glucose Category')
plt.ylabel('Insulin')
plt.legend(title='Diabetes Status', loc='upper right', labels=['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# Comparing the glucose and insulin levels for each individual
for index, row in df.iterrows():
    if row['Glucose'] > row['Insulin']:
        print(f"Individual {index}: Glucose: {row['Glucose']}, Insulin: {row['Insulin']} -> Has diabetes")
    else:
        print(f"Individual {index}: Glucose: {row['Glucose']}, Insulin: {row['Insulin']} -> Is non-diabetic")
```
